## 6. '총구매금액'을 '총방문횟수'로 나누어, '1회방문구매평균'을 계산하고, 생성된 파생변수를 바탕으로 아래와 같이 고객등급을 분류하시오.
- '1회방문구매평균'이 10만원 이상 고객은 A / 5만원 이상은 B / 2만원 이상 C / 나머지는 D 등급으로 분류
- 분류된 고객 등급은 '고객등급'이라는 새로운 파생변수로 생성 

```python
df2['1회방문구매평균'] = df2['총구매금액']/df2['총 방문횟수']

def func6(row):
    if row['1회방문구매평균'] >= 100000:
        return 'A'
    elif row['1회방문구매평균'] >= 50000:
        return 'B'
    elif row['1회방문구매평균'] >= 20000:
        return 'C'
    else:
        return 'D'
df2['고객등급'] = df2.apply(func6, axis=1)
df2
```

```python
# 1회방문구매평균구하기
df2['1회방문구매평균'] = df2['총구매금액'] / df2['총 방문횟수']
df2.head(3)
```

```python
def func8(row):
    if '서울' in row:
        return '서울시'
    elif '용인' in row:
        return '용인시'
    elif '성남' in row:
        return '성남시'
    elif '수원' in row:
        return '수원시'
    elif '광주' in row:
        return '광주'
    else:
        return '기타'

df2['지역(시)'] = df2['주소'].apply(func8)
df2['지역(시)'].value_counts()
```

```python
#노가다 그만 외않되?
list1 = ['서울' , '용인' , '성남', '수원' , '광주', '기타']
def func88(row):
    for i in list1:
        if i in row:
            return i + '시'
    return '기타'

df2['지역(시)'] = df2['주소'].apply(func88)
df2['지역(시)'].value_counts()
```

# 데이터 전처리
- 날짜 , 시간 데이터 기준으로 병합 시 , 시점이 서로 다른 데이터를 병합해야 할때
- 05_Detection : 품질검사 데이터 -> 4시간 마다 수집
- 05_Process_A : 공정 -> 10분 마다 수집

- 어떤 데이터를 기준으로 병합할지 결정후 분석
- 품질데이터 기준 / 공정 데이터 기준 

```python
df_detect = pd.read_csv('05_Detection.csv')
df_process = pd.read_csv('05_Process_A.csv')

df_detect.head() #4시간마다 수집
```

```python
df_process.head() # 10분 마다 수집
```

```python
print(df_detect.shape, df_process.shape) #detect를 기준으로 병합
```

```python
# datetime 타입으로 변환
df_detect['datetime'] = pd.to_datetime(df_detect['date_clean'])
df_process['datetime'] = pd.to_datetime(df_process['DATE'])

                                       
```

```python
# inner join
pd.merge(df_detect, df_process, on = 'datetime' , how ='inner') #근데 왜 174건? micro초 때문에
```

```python
df_detect['datetime'] #micro초 없애야함
```

```python
# 30분 단위로 반올림
df_detect['datetime'].dt.round('30min') #30분 단위로 반올림 ex) 0~30분 => 0분 // 31분 ~ 59분 => 30분
```

```python
# h : 시간
# min or t : 분
# s : 초
# l : 밀리초
# 마이크로초 etc
```

```python
df_detect['datetime_rd'] = df_detect['datetime'].dt.round('30min')
```

```python
df_process.head()
```

```python
df_detect.head()
```

```python
pd.merge(df_detect,df_process,left_on = 'datetime_rd' , right_on = 'datetime' , how ='inner')

#left_on / right_on : 두 테이블의 col이 다를때 키 col을 직접지정
# 값은 공통값이 있어야 함
```

```python
# A1 - A6의 중앙값을 가져와서 merge하자!
p1 = df_process[['datetime' , 'A1', 'A2', 'A3' , 'A4' , 'A5', 'A6']].set_index('datetime')
p1
```

```python
# resample " index가 dateime , 시간 ,데이트 타입 이어야 작동함

p2 = p1.resample('30min').median().reset_index()
p2

# 30q분 간의 데이터 중 중앙값을 가져온다
```

```python
df1 = pd.merge(df_detect, p2, left_on = 'datetime_rd' , right_on = 'datetime')
df1

# datetime_x / datetime_y => x,y 는 같은 이름으로 된 칼럼이 있을때 자동으로 붙음
```

```python
# bfill() -> backward fill
#결측치 이후의 값으로 채우기
df_na['D2TI13029 18'].head(10).bfill()
```

# 이상치 제거
- df.loc[cond1 , 'column'] = value

```python
# 03_Delivery.csv
df1 = pd.read_csv('03_Delivery.csv', encoding = 'cp949')
df1.describe()
#수량에 음수값 있음
# -1 이상치 or 환불한 데이터
```

```python
df1.head()
```

```python
# 음수 이상치 확인
cond1 = (df1['수량'] < 0)
df1.loc[cond1]
```

```python
df1['수량clean'] = df1['수량']
# 이후에 이상치가 아닐 수도 있으니 원본 col을 남겨놓고 새로운 col생산
```

```python
cond1 = df1['수량'] < 0
df1.loc[cond1 , '수량clean'] =1 
df1.describe()
```

```python
# 상품구매금액의 환불한 값을 음수로  추가
df1['상품구매금액clean'] = df1['상품구매금액']
df1.loc[cond1 , '상품구매금액clean'] = -df1['상품구매금액']
# 조건에 해당하는 값들만 음수 붙여서 표현
```

```python
df1['상품구매금액clean'].sum()
```

```python
df1.describe()
```

```python
# 차이
df1['상품구매금액'].sum() - df1['상품구매금액clean'].sum()
```

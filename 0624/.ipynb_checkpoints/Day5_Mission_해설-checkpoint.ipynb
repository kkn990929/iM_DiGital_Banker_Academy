## 5. 각 '자재명' 별 '쿠팡출고수량', '입고수량'의 평균을 계산하고, 계산된 결과에서 '쿠팡출고수량'의 상위 10% '자재명'을 'A' / 상위 30%를 'B' / 상위 60%를 'C' / 나머지를 'D'로 '등급분류'라는 파생변수를 생성하고, 각 등급에서 가장 많이 등장한 '카테고리'항목이 무엇인지 확인하시오. 

```python
#처음부터 카테고리 추가
p3 = df1.pivot_table(index = ['자재명', '카테고리'] , values = ['쿠팡출고수량' , '입고수량'] , aggfunc = 'mean').reset_index()

p3
```

```python
cond1 = (p3['쿠팡출고수량'] >= p3['쿠팡출고수량'].quantile(0.4))
cond2 = (p3['쿠팡출고수량'] >= p3['쿠팡출고수량'].quantile(0.7))
cond3 = (p3['쿠팡출고수량'] >= p3['쿠팡출고수량'].quantile(0.9))

p3['등급'] = 'D'

p3.loc[cond1, '등급' ] = 'C'
p3.loc[cond2, '등급' ] = 'B'
p3.loc[cond3, '등급' ] = 'A'
# D를 기초로 하나씩 조건 수행
```

```python
p3['등급'].value_counts()
```

```python
# 각 등급에서 가장 많이 등장한 '카테고리'항목이 무엇인지 확인하시오
pd.crosstab(p3['등급'] , p3['카테고리'])
```

```python
df['주문일시dt'] = pd.to_datetime(df['주문일시dt'])
df = df.sort_values(by = '주문일시dt')
df['구매주기'] = df.groupby('회원번호')['주문일시dt'].diff().dt.days # 정렬 후 회원별로 날짜 차이 계산
df['구매주기'] = np.where(df['구매주기'] == 0, pd.NA, df['구매주기']) # mean에 넣지 않기 위해 nan값으로 대치
# df['평균구매주기'] = df.groupby('회원번호')['구매주기'].transform('mean') # 평균구매주기
# df['평균구매주기'] = df['평균구매주기'].round() # 반올림
df['방문수'] = df.groupby('회원번호')['구매주기'].transform('count') # 방문수 구하기
```

```python
df.columns
```

```python
df['Customer_Type'].value_counts()
```

```python
df['Customer_Type1'] = 'other'
df.loc[df['Customer_Type'].isin(['우수고객']),'Customer_Type1'] = '우수고객'

```

```python
# "other"과 "초우량고객"만 필터링
filtered_df = df[df['Customer_Type1'].isin(['우수고객', 'other'])]

# 각 고객 유형에 따른 제품군 비율 계산
product_ratios = (
    filtered_df.groupby(['Customer_Type1', '제품군'])
    .size()
    .unstack(fill_value=0)
    .apply(lambda x: (x / x.sum()).round(3), axis=1)
    .apply(lambda x: x.sort_values(ascending=False), axis=1)  # 내림차순 정렬
)

# DataFrame 형태로 결과 출력
print("제품군 비율 (DataFrame 형태):")
product_ratios
```

```python
# 다중 막대 그래프 생성
fig, ax = plt.subplots(figsize=(10, 6))

# 제품군 및 고객 유형 리스트
groups = product_ratios.columns
types = product_ratios.index

# 막대의 x 좌표 설정
bar_width = 0.35
indices = np.arange(len(groups))

# 각 고객 유형의 막대 그래프
for i, customer_type in enumerate(types):
    ax.bar(indices + i * bar_width, product_ratios.loc[customer_type], bar_width, label=customer_type)

# 그래프 설정
ax.set_title('other과 초우량고객의 제품군 비교')
ax.set_xlabel('제품군')
ax.set_ylabel('비율')
ax.set_xticks(indices + bar_width / 2)
ax.set_xticklabels(groups)
ax.legend(title='제품군')
plt.xticks(rotation=35)
plt.tight_layout()

# 그래프 보여주기
plt.show()
```

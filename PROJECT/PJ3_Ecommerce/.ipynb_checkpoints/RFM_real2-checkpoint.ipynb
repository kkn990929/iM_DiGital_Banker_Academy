```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
matplotlib.rc('font', family = 'Malgun Gothic')
matplotlib.rc('axes', unicode_minus=False)
warnings.filterwarnings('ignore')
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('data_total.csv')
df_member = pd.read_csv('member_data.csv', encoding = 'cp949')
df.columns
```

```python
df_member
```

```python
df['주문일시dt'] = pd.to_datetime(df['주문일시dt'])
df = df.sort_values(by = '주문일시dt')
df['구매주기'] = df.groupby('회원번호')['주문일시dt'].diff().dt.days # 정렬 후 회원별로 날짜 차이 계산
df['구매주기'] = np.where(df['구매주기'] == 0, pd.NA, df['구매주기']) # mean에 넣지 않기 위해 nan값으로 대치
# df['평균구매주기'] = df.groupby('회원번호')['구매주기'].transform('mean') # 평균구매주기
# df['평균구매주기'] = df['평균구매주기'].round() # 반올림
df['방문수'] = df.groupby('회원번호')['구매주기'].transform('count') # 방문수 구하기
```

```python
df['방문수'].value_counts()
```

```python
df = df[df['주문취소여부'] != '주문취소']
# '주문일시' 컬럼을 datetime 형식으로 변환
df['주문일시'] = pd.to_datetime(df['주문일시'])

# 2. 데이터 전처리 (결측치 및 중복값 처리)
# df.drop_duplicates(inplace=True)

# 3. RFM 값 계산

# 기준 날짜 설정 (데이터셋의 마지막 날짜로 설정)
current_date = df['주문일시'].max()  # 데이터셋의 마지막 날짜

# Recency 계산
rfm_recency = df.groupby('회원번호').agg({
    '주문일시': lambda x: (current_date - x.max()).days
}).rename(columns={'주문일시': 'Recency'})

# Frequency 계산
rfm_frequency = df.groupby('회원번호').agg({
    '주문일시': 'count'
}).rename(columns={'주문일시': 'Frequency'})

# Monetary 계산
rfm_monetary = df.groupby('회원번호').agg({
    '구매금액': 'sum'
}).rename(columns={'구매금액': 'Monetary'})

# RFM 테이블 병합
df = rfm_recency.join(rfm_frequency).join(rfm_monetary)

# 4. RFM 점수 계산

# Recency, Frequency, Monetary에 대한 점수 매기기 (1부터 5까지)
df['R'] = pd.qcut(df['Recency'], 5, labels=range(5, 0, -1))
df['F'] = pd.qcut(df['Frequency'], 5, labels=range(1, 6))
df['M'] = pd.qcut(df['Monetary'], 5, labels=range(1, 6))

# RFM 점수 합산
df['RFM_Score'] = df[['R', 'F', 'M']].sum(axis=1)

# 5. 고객 세그먼트 생성
    
# def segment_customer(df):
#     if df['RFM_Score'] >= 9:
#         return '우수고객'
#     elif df['RFM_Score'] >= 6:
#         return '일반고객'
#     else:
#         return '잠재고객'

# df['Customer_Segment'] = df.apply(segment_customer, axis=1)

# 고객 세그먼트 생성
def segment_customer(df):
    if df['R'] >= 4 and df['F'] >= 4 and df['M'] >= 4:
        return '초우량고객'
    elif df['R'] >= 4 and df['F'] >= 4 and df['M'] < 4:
        return '저수익고객'
    elif df['R'] >= 4 and df['M'] >= 4 and df['F'] < 4:
        return '신규고객'
    elif df['F'] >= 4 and df['M'] >= 4 and df['R'] < 4:
        return '이탈고객'
    else:
        return '일반고객'
    # elif df['RFM_Score'] >= 9:
    #     return '우수고객'
    # elif df['RFM_Score'] >= 6:
    #     return '일반고객'
    # else:
    #     return '잠재고객'

df['Customer_Type'] = df.apply(segment_customer, axis=1)

# 회원번호를 컬럼으로 추가
df.reset_index(inplace=True)

# 결과 출력
# print(df)
df['Recency']

```

```python
import matplotlib.pyplot as plt

# 고객 유형과 각 유형의 수
customer_types = {
    '일반고객': 7563,
    '초우량고객': 3510,
    '이탈고객': 1093,
    '저수익고객': 191,
    '신규고객': 136
}

# 전체 고객 수 계산
total_customers = sum(customer_types.values())

# 각 유형의 비율 계산
ratios = [count / total_customers for count in customer_types.values()]

# 초록 계열 색상 목록
colors = ['#2E8B57', '#3CB371', '#66CDAA', '#8FBC8F', '#20B2AA']

# 파이 차트 그리기
plt.figure(figsize=(15, 8))
wedges, texts, autotexts = plt.pie(
    ratios, 
    labels=customer_types.keys(), 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=colors
)

# 범례 추가
plt.legend(
    wedges, 
    customer_types.keys(),
    title="고객 유형",
    loc="upper right",  # 범례 위치 설정
    bbox_to_anchor=(1, 1, 0.3, 1)  # 범례가 그래프 내부에 있게 함 (x, y, width, height)
)

# 제목 및 비율 표기 형식
plt.title('고객 유형별 비율')
plt.setp(autotexts, size=5, weight="bold", color="white")

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()  # 레이아웃 조정

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
matplotlib.rc('font', family = 'Malgun Gothic')
matplotlib.rc('axes', unicode_minus=False)
warnings.filterwarnings('ignore')
import matplotlib.ticker as ticker
```

```python
data_sm = pd.read_csv('data_sm.csv')
data_sm
```

```python
# '주문일시' 컬럼을 datetime 형식으로 변환
data_sm['주문일시'] = pd.to_datetime(data_sm['주문일시'])

# 월 추출
data_sm['월'] = data_sm['주문일시'].dt.to_period('M')

# 월별 회원별 구매 횟수 및 총액 집계
monthly_summary = data_sm.groupby(['월', '회원번호']).agg({
    '구매금액': 'sum',
    '주문일시': 'count'
}).reset_index()
monthly_summary.rename(columns={'주문일시': '구매횟수'}, inplace=True)

# 구매 횟수와 총액 조건 필터링
filtered_members = monthly_summary[
    (monthly_summary['구매횟수'] >= 5) & (monthly_summary['구매금액'] >= 200000)
]

# 결과를 데이터프레임으로 변환
result_df = filtered_members[['회원번호', '월', '구매횟수', '구매금액']].reset_index(drop=True)
result_df
```

```python
# '주문일시' 컬럼을 datetime 형식으로 변환
data_sm['주문일시'] = pd.to_datetime(data_sm['주문일시'])

# 월 추출
data_sm['월'] = data_sm['주문일시'].dt.to_period('M')

# 월별 회원별 구매 횟수 및 총액 집계
monthly_summary = data_sm.groupby(['월', '회원번호']).agg({
    '구매금액': 'sum',
    '주문일시': 'count'
}).reset_index()
monthly_summary.rename(columns={'주문일시': '구매횟수'}, inplace=True)

# 월별로 회원번호의 개수 카운트
monthly_member_count = monthly_summary.groupby('월')['회원번호'].nunique().reset_index()
monthly_member_count.rename(columns={'회원번호': '회원수'}, inplace=True)

# 월별 회원별 집계 데이터프레임과 회원수 데이터프레임을 병합
result_df = pd.merge(monthly_summary, monthly_member_count, on='월')

# 조건 필터링
filtered_members = result_df[
    (result_df['구매횟수'] >= 5) & (result_df['구매금액'] >= 200000)
]

# 결과를 데이터프레임으로 변환
result_df = filtered_members[['회원번호', '월', '구매횟수', '구매금액', '회원수']].reset_index(drop=True)
result_df['회원수'].unique()
result_df[result_df['월'] == '2022-01']
```

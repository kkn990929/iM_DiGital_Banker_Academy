## 검정

```python
status_counts= df_high['회원상태'].value_counts()
```

```python
# 빈도 테이블을 기반으로 관찰된 값 준비
observed = pd.DataFrame({
    '회원상태': status_counts.index,
    '빈도': status_counts.values
})

# 카이제곱 검정 수행
chi2, p, dof, expected = chi2_contingency([observed['빈도']])

print(f"Chi2 Statistic: {chi2}")
print(f"P-value: {p}")
# 두개는 독립적이다(차이가 없다)
```

```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
matplotlib.rc('font', family = 'Malgun Gothic')
matplotlib.rc('axes', unicode_minus=False)
warnings.filterwarnings('ignore')
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('data_total.csv')
df
```

```python
df['주문일시dt'] = pd.to_datetime(df['주문일시dt'])
df = df.sort_values(by = '주문일시dt')
df['구매주기'] = df.groupby('회원번호')['주문일시dt'].diff().dt.days # 정렬 후 회원별로 날짜 차이 계산
df['구매주기'] = np.where(df['구매주기'] == 0, pd.NA, df['구매주기']) # mean에 넣지 않기 위해 nan값으로 대치
# df['평균구매주기'] = df.groupby('회원번호')['구매주기'].transform('mean') # 평균구매주기
# df['평균구매주기'] = df['평균구매주기'].round() # 반올림
df['방문수'] = df.groupby('회원번호')['구매주기'].transform('count') # 방문수 구하기
```

```python
df['방문수'].value_counts()
```

```python
 # pd.DataFrame(data)

# '주문일시' 컬럼을 datetime 형식으로 변환
df['주문일시'] = pd.to_datetime(df['주문일시'])

# 데이터 전처리 (중복값 제거)
df.drop_duplicates(inplace=True)

# 기준 날짜 설정 (데이터셋의 마지막 날짜로 설정)
current_date = df['주문일시'].max()

# Recency 계산
rfm_recency = df.groupby('회원번호').agg({
    '주문일시': lambda x: (current_date - x.max()).days
}).rename(columns={'주문일시': 'Recency'})

# Frequency 계산
rfm_frequency = df.groupby('회원번호').agg({
    '주문일시': 'count'
}).rename(columns={'주문일시': 'Frequency'})

# Monetary 계산
rfm_monetary = df.groupby('회원번호').agg({
    '구매금액': 'sum'
}).rename(columns={'구매금액': 'Monetary'})

# RFM 테이블 병합
rfm_table = rfm_recency.join(rfm_frequency).join(rfm_monetary)

# Recency, Frequency, Monetary에 대한 점수 매기기 (1부터 5까지)
rfm_table['R'] = pd.qcut(rfm_table['Recency'], 5, labels=range(5, 0, -1)).astype(int)
rfm_table['F'] = pd.qcut(rfm_table['Frequency'], 5, labels=range(1, 6)).astype(int)
rfm_table['M'] = pd.qcut(rfm_table['Monetary'], 5, labels=range(1, 6)).astype(int)


# RFM 점수 합산 (가중치 적용)
# Recency에 40%, Frequency에 20%, Monetary에 40%의 가중치 적용
rfm_table['RFM_Score'] = (
    rfm_table['R'] * 0.4 +
    rfm_table['F'] * 0.2 +
    rfm_table['M'] * 0.4
)

# 고객 세그먼트 생성
def segment_customer(rfm_table):
    # if rfm_table['R'] >= 4 and rfm_table['F'] >= 4 and rfm_table['M'] >= 4:
    #     return '초우량고객'
    # elif rfm_table['R'] >= 4 and rfm_table['F'] >= 4 and rfm_table['M'] < 4:
    #     return '저수익고객'
    # elif rfm_table['R'] >= 4 and rfm_table['M'] >= 4 and rfm_table['F'] < 4:
    #     return '신규고객'
    # elif rfm_table['F'] >= 4 and rfm_table['M'] >= 4 and rfm_table['R'] < 4:
    #     return '이탈고객'
    if rfm_table['RFM_Score'] >= 4.2:
        return '우수고객'
    elif rfm_table['RFM_Score'] >= 6:
        return '일반고객'
    else:
        return '잠재고객'

rfm_table['Customer_Type'] = rfm_table.apply(segment_customer, axis=1)

# 회원번호를 컬럼으로 추가
rfm_table.reset_index(inplace=True)

# 결과 출력
rfm_table['RFM_Score'].describe()
```

```python
# Count of each customer type
customer_type_counts = rfm_table['Customer_Type'].value_counts().reset_index()
customer_type_counts.columns = ['Customer_Type', 'Count']

# Bar plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=customer_type_counts, x='Customer_Type', y='Count', palette='viridis')
plt.title('RFM분석결과', fontsize=16)
plt.xlabel('Customer_Type', fontsize=12)
plt.ylabel('고객 수', fontsize=12)
# plt.xticks(rotation=45)
plt.show()
```

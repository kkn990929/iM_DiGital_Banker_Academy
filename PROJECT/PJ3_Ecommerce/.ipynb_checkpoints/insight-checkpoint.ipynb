```python
# 우량고객 데이터만 추출
vip_customers = df[df['Customer_Type'] == '초우량고객']

# VIP 고객의 회원상태별 빈도 계산
vip_address_counts = vip_customers['주문요일'].value_counts().reset_index()
vip_address_counts.columns = ['주문요일', '고객수']

# 시각화
plt.figure(figsize=(15, 6))
sns.barplot(x='주문요일', y='고객수', data=vip_address_counts, palette='viridis')

# 각 막대 위에 고객 수 표시
for index, row in vip_address_counts.iterrows():
    plt.text(index, row['고객수'] + 0.1, f'{row["고객수"]}', 
             color='black', ha="center", va='bottom')

# 플롯 설정
plt.title('우량 고객의 주문요일 분포')
plt.xlabel('주문요일')
plt.ylabel('고객수')
plt.ylim(0, vip_address_counts['고객수'].max() + 20000)  # y축 범위 설정
plt.xticks(rotation=15)  # x축 레이블 회전
plt.show()
```

```python
a = pd.DataFrame(df[df['Customer_Type'] == '초우량고객']['제품군'].value_counts()) 
b = pd.DataFrame(df[df['Customer_Type'] == '이탈고객']['제품군'].value_counts()) 
df_a = a.div(a.sum(axis=0), axis = 1)
df_a['제품군차이'] = b.div(b.sum(axis=0), axis = 1)
df_a['차이비율'] = (df_a['제품군차이'] - df_a['count']) / df_a['count'] * 100
df_a['차이'] = df_a['제품군차이'] - df_a['count']
df_a.sort_values(by='차이', ascending=False) 
```

```python
# "이탈고객"과 "초우량고객"만 필터링
filtered_df = df[df['Customer_Type'].isin(['이탈고객', '초우량고객'])]

# 각 고객 유형에 따른 제품군 비율 계산
product_ratios = (
    filtered_df.groupby(['Customer_Type', '제품군'])
    .size()
    .unstack(fill_value=0)
    .apply(lambda x: x / x.sum(), axis=1)
)

# DataFrame 형태로 결과 출력
print("제품군 비율 (DataFrame 형태):")
product_ratios
```

```python
# "이탈고객"과 "초우량고객"만 필터링
filtered_df = df[df['Customer_Type'].isin(['이탈고객', '초우량고객'])]

# 각 고객 유형에 따른 제품군 비율 계산
product_ratios = (
    filtered_df.groupby(['Customer_Type', '나이대'])
    .size()
    .unstack(fill_value=0)
    .apply(lambda x: x / x.sum(), axis=1)
)

# DataFrame 형태로 결과 출력
print("제품군 비율 (DataFrame 형태):")
product_ratios
```

```python
# 다중 막대 그래프 생성
fig, ax = plt.subplots(figsize=(10, 6))

# 나이대 및 고객 유형 리스트
age_groups = product_ratios.columns
customer_types = product_ratios.index

# 막대의 x 좌표 설정
bar_width = 0.35
indices = np.arange(len(age_groups))

# 각 고객 유형의 막대 그래프
for i, customer_type in enumerate(customer_types):
    ax.bar(indices + i * bar_width, product_ratios.loc[customer_type], bar_width, label=customer_type)

# 그래프 설정
ax.set_title('나이대 비율 비교 (이탈고객 vs 초우량고객)')
ax.set_xlabel('나이대')
ax.set_ylabel('비율')
ax.set_xticks(indices + bar_width / 2)
ax.set_xticklabels(age_groups)
ax.legend(title='고객 유형')
plt.xticks(rotation=0)
plt.tight_layout()

# 그래프 보여주기
plt.show()
```

```python
filtered_df = df[df['Customer_Type'].isin(['초우량고객',])]

# 각 고객 유형에 따른 제품군 비율 계산 후 내림차순 정렬
product_ratios = (
    filtered_df.groupby(['Customer_Type', '등록카드'])
    .size()
    .unstack(fill_value=0)
    .apply(lambda x: (x / x.sum()).round(3), axis=1)
    .apply(lambda x: x.sort_values(ascending=False), axis=1)  # 내림차순 정렬
)

# 다중 막대 그래프 생성
fig, ax = plt.subplots(figsize=(12, 8))

# 제품군 및 고객 유형 리스트
product_groups = product_ratios.columns
customer_types = product_ratios.index

# 막대의 x 좌표 설정
bar_width = 0.3
indices = np.arange(len(product_groups))

# 연한 초록 계열 색상 목록
colors = ['#98FB98', '#90EE90', '#8FBC8F', '#3CB371', '#2E8B57']

# 각 고객 유형의 막대 그래프
for i, customer_type in enumerate(customer_types):
    ax.bar(indices + i * bar_width, product_ratios.loc[customer_type], bar_width, label=customer_type, color=colors[i % len(colors)])

# 그래프 설정
ax.set_title('초우량고객별 등록카드', fontsize=15)
ax.set_xlabel('등록카드', fontsize=12)
ax.set_ylabel('비율', fontsize=12)
ax.set_xticks(indices + bar_width / 2)
ax.set_xticklabels(product_groups)
ax.legend(title='고객 유형')
plt.xticks(rotation=35)
plt.tight_layout()

# 그래프 보여주기
plt.show()
```

```python
df['주문일시dt'] = pd.to_datetime(df['주문일시dt'])
df = df.sort_values(by = '주문일시dt')
df['구매주기'] = df.groupby('회원번호')['주문일시dt'].diff().dt.days # 정렬 후 회원별로 날짜 차이 계산
df['구매주기'] = np.where(df['구매주기'] == 0, pd.NA, df['구매주기']) # mean에 넣지 않기 위해 nan값으로 대치
# df['평균구매주기'] = df.groupby('회원번호')['구매주기'].transform('mean') # 평균구매주기
# df['평균구매주기'] = df['평균구매주기'].round() # 반올림
df['방문수'] = df.groupby('회원번호')['구매주기'].transform('count') # 방문수 구하기
```

```python
# '주문일시' 컬럼을 datetime 형식으로 변환
df['주문일시'] = pd.to_datetime(df['주문일시'])

# 2. 데이터 전처리 (결측치 및 중복값 처리)
df.drop_duplicates(inplace=True)

# 3. RFM 값 계산

# 기준 날짜 설정 (데이터셋의 마지막 날짜로 설정)
current_date = df['주문일시'].max()  # 데이터셋의 마지막 날짜

# Recency 계산
rfm_recency = df.groupby('회원번호').agg({
    '주문일시': lambda x: (current_date - x.max()).days
}).rename(columns={'주문일시': 'Recency'})

# Frequency 계산
rfm_frequency = df.groupby('회원번호').agg({
    '주문일시': 'count'
}).rename(columns={'주문일시': 'Frequency'})

# Monetary 계산
rfm_monetary = df.groupby('회원번호').agg({
    '구매금액': 'sum'
}).rename(columns={'구매금액': 'Monetary'})


# RFM 테이블 생성
df = rfm_recency.join(rfm_frequency).join(rfm_monetary)

# Recency, Frequency, Monetary에 대한 점수 매기기 (1부터 5까지)
df['R'] = pd.qcut(df['Recency'], 5, labels=range(5, 0, -1)).astype(int)
df['F'] = pd.qcut(df['Frequency'], 5, labels=range(1, 6)).astype(int)
df['M'] = pd.qcut(df['Monetary'], 5, labels=range(1, 6)).astype(int)

# RFM 점수 합산 (가중치 적용)
# Recency에 40%, Frequency에 30%, Monetary에 40%의 가중치 적용
df['RFM_Score'] = (
    df['R'] * 0.4 +
    df['F'] * 0.2 +
    df['M'] * 0.4
)
# 
# RFM_Score의 총점을 정수로 반올림
df['RFM_Score'] = df['RFM_Score'].round(2)

# # 고객 세그먼트 생성
# def segment_customer(df):
#     if df['RFM_Score'] >= 4.5:  # RFM 점수가 4.5 이상인 고객을 우수 고객으로 분류
#         return '우수고객'
#     elif df['RFM_Score'] >= 3.0:  # RFM 점수가 3.0 이상인 고객을 일반 고객으로 분류
#         return '일반고객'
#     else:
#         return '잠재고객'

df['Customer_Segment'] = df.apply(segment_customer, axis=1)

# 고객 유형 세그먼트 생성
def segment_customer_type(df):
    if df['R'] >= 4 and df['F'] >= 4 and df['M'] >= 4:
        return '초우량고객'
    elif df['R'] >= 4 and df['F'] >= 4 and df['M'] < 4:
        return '저수익고객'
    elif df['R'] >= 4 and df['M'] >= 4 and df['F'] < 4:
        return '신규고객'
    elif df['F'] >= 4 and df['M'] >= 4 and df['R'] < 4:
        return '이탈고객'
    else:
        return '일반고객'

df['Customer_Type'] = df.apply(segment_customer_type, axis=1)

# 회원번호를 컬럼으로 추가
df.reset_index(inplace=True)

# 결과 출력
df['Customer_Type'].value_counts()

```

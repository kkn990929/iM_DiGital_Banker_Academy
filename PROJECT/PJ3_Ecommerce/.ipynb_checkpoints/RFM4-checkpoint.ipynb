```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
matplotlib.rc('font', family = 'Malgun Gothic')
matplotlib.rc('axes', unicode_minus=False)
warnings.filterwarnings('ignore')
import matplotlib.ticker as ticker
```

```python
df_total = pd.read_csv('data_total.csv')
df_total
```

```python
df_score = pd.read_csv('data_RF_Score.csv')
df_score


# 조건 정의
cond1 = df_score['월간구매금액기준구분'] == 'HIGH'
cond2 = df_score['월간주문횟수기준구분'] == 'LOW'

# 두 조건을 AND로 결합하여 행 선택
result = df_score.loc[cond1 & cond2]

# 결과 출력
result
```

```python
# 조건 정의
cond1 = df_score['월간구매금액기준구분'] == 'LOW'
cond2 = df_score['월간주문횟수기준구분'] == 'HIGH'

# 두 조건을 AND로 결합하여 행 선택
result1 = df_score.loc[cond1 & cond2]

# 결과 출력
result1
```

```python
# 조건별 개수 계산
cond1_count = len(df_score[(df_score['월간구매금액기준구분'] == 'HIGH') & (df_score['월간주문횟수기준구분'] == 'LOW')])
cond2_count = len(df_score[(df_score['월간구매금액기준구분'] == 'LOW') & (df_score['월간주문횟수기준구분'] == 'HIGH')])
cond3_count = len(df_score[(df_score['월간구매금액기준구분'] == 'HIGH') & (df_score['월간주문횟수기준구분'] == 'HIGH')])
cond4_count = len(df_score[(df_score['월간구매금액기준구분'] == 'LOW') & (df_score['월간주문횟수기준구분'] == 'LOW')])

# 조건별 카운트 데이터프레임 생성
condition_counts = pd.DataFrame({
    '조건': ['HIGH 금액 & LOW 주문횟수', 
             'LOW 금액 & HIGH 주문횟수', 
             'HIGH 금액 & HIGH 주문횟수', 
             'LOW 금액 & LOW 주문횟수'],  # 잘못된 중복 라벨 수정
    '행 개수': [cond1_count, cond2_count, cond3_count, cond4_count]
})
# 내림차순으로 정렬
condition_counts = condition_counts.sort_values(by='행 개수', ascending=False).reset_index(drop=True)
# 시각화
plt.figure(figsize=(12, 6))  # 적절한 가로세로 비율로 크기 조정
bar_plot = sns.barplot(x='조건', y='행 개수', data=condition_counts, palette='viridis')

# 각 막대 위에 행 개수 표시
for index, row in condition_counts.iterrows():
    bar_plot.text(index, row['행 개수'] + 500, f'{row["행 개수"]:,}',  # 천단위 구분자 추가
                  color='black', ha="center", va='bottom')

# 플롯 설정
plt.title('조건별 행 개수')
plt.xlabel('조건')
plt.ylabel('행 개수')
plt.ylim(0, condition_counts['행 개수'].max() + 5000)  # 최대값 기준으로 y축 범위 조정
# plt.xticks(rotation=15)  # x축 레이블 회전으로 가독성 향상
plt.show()
```

```python
# df_10 데이터랑 df_10_F랑 병합
df = pd.merge(df_total, df_score, on='회원번호', how='right')
```

```python
df['score_sum'].value_counts()
```

```python
# score_sum의 빈도수 계산 및 내림차순 정렬
score_order = df['score_sum'].value_counts().index

# score_sum에 대한 빈도수 시각화
plt.figure(figsize=(10, 6))
sns.countplot(x='score_sum', data=df, order=score_order)
plt.title('score_sum 빈도수')
plt.xlabel('score_sum')
plt.ylabel('빈도수')
plt.show()
```

```python
# score_sum에 따라 LOW와 HIGH로 분류하는 함수 정의
def categorize_score(score):
    if 2 <= score <= 5:
        return '일반'
    elif 6 <= score <= 8:
        return 'VIP'
    else:
        return 'UNKNOWN'  # 예외 처리를 위해 추가

# 새로운 컬럼 'Category' 추가하여 LOW와 HIGH로 분류
df_score['Category'] = df_score['score_sum'].apply(categorize_score)

# 결과 출력
df_score['Category'].value_counts()
```

```python
df_score.to_csv('data_RF_Score.csv')
```

```python
# score_sum에 따라 LOW와 HIGH로 분류하는 함수 정의
def categorize_score(score):
    if 2 <= score <= 6:
        return '일반'
    elif 7 <= score <= 8:
        return 'VIP'
    else:
        return 'UNKNOWN'  # 예외 처리를 위해 추가

# 새로운 컬럼 'Category' 추가하여 LOW와 HIGH로 분류
df_score['Category'] = df_score['score_sum'].apply(categorize_score)

# 결과 출력
df_score['Category'].value_counts()
```

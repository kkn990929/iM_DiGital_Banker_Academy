```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
matplotlib.rc('font', family = 'Malgun Gothic')
matplotlib.rc('axes', unicode_minus=False)
warnings.filterwarnings('ignore')
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('data_total.csv')
df
```

```python
# 주문 취소 행 삭제
df = df[df['주문취소여부'] != '주문취소']
```

```python
df_10 = df[df['주문월'] == 10]
```

```python
df_10
```

```python
df_total = pd.read_csv('data_total.csv')
df_total.head()
```

```python
# 주문월에 10이 있는지 여부에 따라 점수 부여 함수 정의
def assign_score(group):
    if 10 in group['주문월'].values:
        return 3
    else:
        return 0

# 회원번호별로 그룹화하여 점수 부여
df['점수'] = df.groupby('회원번호').apply(assign_score).reset_index(drop=True)

# 결과 출력
print(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
matplotlib.rc('font', family = 'Malgun Gothic')
matplotlib.rc('axes', unicode_minus=False)
warnings.filterwarnings('ignore')
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('data_RF_Score_ver2.csv')
df
```

```python
df['Category'].value_counts()
```

```python
sns.countplot(x = 'Category', data = df)
```

# VIP 비교

```python
# VIP 고객 데이터만 추출
vip_customers = df[df['Category'] == 'VIP']

# VIP 고객의 주소지별 빈도 계산
vip_address_counts = vip_customers['주소지'].value_counts().reset_index()
vip_address_counts.columns = ['주소지', '고객수']

# 시각화
plt.figure(figsize=(15, 6))
sns.barplot(x='주소지', y='고객수', data=vip_address_counts, palette='viridis')

# 각 막대 위에 고객 수 표시
for index, row in vip_address_counts.iterrows():
    plt.text(index, row['고객수'] + 0.1, f'{row["고객수"]}', 
             color='black', ha="center", va='bottom')

# 플롯 설정
plt.title('VIP 고객의 주소지 분포')
plt.xlabel('주소지')
plt.ylabel('고객수')
plt.ylim(0, vip_address_counts['고객수'].max() + 1)  # y축 범위 설정
plt.xticks(rotation=15)  # x축 레이블 회전
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
matplotlib.rc('font', family = 'Malgun Gothic')
matplotlib.rc('axes', unicode_minus=False)
warnings.filterwarnings('ignore')
import matplotlib.ticker as ticker
```

# 점수를 매겨보자

## Monetary_score

```python
df = pd.read_csv('data_total.csv')
df
```

```python
# 주문 취소 행 삭제
df = df[df['주문취소여부'] != '주문취소']
```

```python
df_10 = df[df['주문월'] == 10]
df_10.to_csv('df_month10.csv')
```

```python
df_10 = pd.read_csv('data_RF.csv')
df_10['월간구매금액'].value_counts()
```

```python
# 월간구매금액의 분포를 보자
df_10['월간구매금액'].describe().round(0)
```

```python
# 박스플롯 생성
plt.figure(figsize=(15, 8))
plt.boxplot(df_10['월간구매금액'])

# y축 레이블 포맷 설정
current_values = plt.gca().get_yticks()  # 현재 y축의 tick 위치를 가져옵니다
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])

plt.title('월간 구매 금액의 박스플롯')
plt.xlabel('월간구매금액')
plt.ylabel('금액')
plt.show()
```

```python
# 백분위수 값 계산
p25 = df_10['월간구매금액'].quantile(0.25)
p50 = df_10['월간구매금액'].quantile(0.50)
p75 = df_10['월간구매금액'].quantile(0.75)

# 점수 부여 함수 정의
def assign_score(value):
    if value <= p25:
        return 1
    elif value <= p50:
        return 2
    elif value <= p75:
        return 3
    else:
        return 4

# 새로운 칼럼 추가
df_10['Monatery점수'] = df_10['월간구매금액'].apply(assign_score)

# 결과 출력
df_10
```

```python
p25, p50, p75
```

```python
# df_10 = df_10.drop(columns = '점수')
df_10.rename(columns={'Monatery점수': 'Monetary_score'}, inplace=True)
```

```python
df_10['Monetary_score'].value_counts()
```

## 구매횟수(Frequency)

```python
df_10['월간주문횟수'].value_counts()
```

```python
# 박스플롯 생성
plt.figure(figsize=(15, 8))
plt.boxplot(df_10['월간주문횟수'])

# y축 레이블 포맷 설정
current_values = plt.gca().get_yticks()  # 현재 y축의 tick 위치를 가져옵니다
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])

plt.title('월간 주문 횟수의 박스플롯')
plt.xlabel('월간주문횟수')
plt.ylabel('주문횟수')
plt.show()
```

```python
df_10['월간주문횟수'].describe().round(2)
```

```python
# 백분위수 값 계산
p25 = df_10['월간주문횟수'].quantile(0.25)
p50 = df_10['월간주문횟수'].quantile(0.50)
p75 = df_10['월간주문횟수'].quantile(0.75)

# 점수 부여 함수 정의
def assign_score(value):
    if value <= p25:
        return 1
    elif value <= p50:
        return 2
    elif value <= p75:
        return 3
    else:
        return 4

# 새로운 칼럼 추가
df_10['Frequency_score'] = df_10['월간주문횟수'].apply(assign_score)

# 결과 출력
df_10
```

```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
matplotlib.rc('font', family = 'Malgun Gothic')
matplotlib.rc('axes', unicode_minus=False)
warnings.filterwarnings('ignore')
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('data_total.csv')
df
```

## 10월 자료만 뽑아서 기준만들자

```python
# 주문 취소 행 삭제
df = df[df['주문취소여부'] != '주문취소']
```

```python
df['주문취소여부'].value_counts()
```

```python
df.info()
```

```python
# 주문월 칼럼 만들기
# df['주문월'] = df['주문일시dt'].dt.month
```

```python
df_10 = df[df['주문월'] == 10]
```

```python
df_10['회원번호'].value_counts()
```

```python
# 10월 주문만 추출
df_10 = df[df['주문월'] == 10]
```

```python
# 10월 주문 내 회원번호별 구매금액 구하기
df_10 = df_10.groupby('회원번호')['구매금액'].sum().reset_index().sort_values(by = '구매금액' , ascending = False)
df_10
```

```python
df_10['구매금액'].describe().round(0)
```

```python
df_10[df_10['구매금액'] >= 90000] # 상위 25%가 적당하다
```

# Monetary

## 9만원으로 구분하자
- 근데 10월기준으로 하면 나머지 월은 어쩌지
- 10월과 11월에 구매이력이 있는 사람은
- 10월 데이터만 가지고 놀자

```python
df_10 = df[df['주문월'] == 10]
df_10 = df_10.drop(columns = 'Monetary')
df_10
```

```python
# 10월에 회원번호 별 구매금액 구하기
df_10['회원번호'].value_counts()
```

```python
df_10_1 = df_10.groupby('회원번호')['구매금액'].sum().reset_index()
```

```python
# 10월만 뽑은 데이터에 월간구매금액 구하기
df_10 = pd.merge(df_10, df_10_1, on='회원번호', how='inner')
```

```python
# 월간구매금액 col이름 바꾸기
df_10.rename(columns = {'구매금액_y' : '월간구매금액'}, inplace = True)
```

```python
# 월간구매금액 기준으로 월간구매금액기준에 HIGH/ LOW만들기
df_10['월간구매금액기준구분'] = 'LOW'
df_10.loc[df_10['월간구매금액'] >= 90000 , '월간구매금액기준구분'] = 'HIGH'
```

```python
df_10['월간구매금액기준구분'].value_counts()
```

# Frequency
- 10월에 얼마나 구매했는가
- 회원번호 갯수를 카운트?
- 행 갯수가 곧 주문갯수

```python
# 회원번호별 갯수를 카운트 해보자
df_10_F = df_10.groupby('회원번호').size().reset_index(name='주문횟수').sort_values(by = '주문횟수', ascending = False)
# 167건 주문은 뭐여
```

```python
# 주문횟수 통계보자
df_10_F['주문횟수'].describe().round(0)
```

```python
# 기준 8로 잡아보자 몇개임?
df_10_F[df_10_F['주문횟수'] >= 8]
# 5570개 중 1571개 약 상위 28%
```

## 8회로 구분하자

```python
# df_10 데이터랑 df_10_F랑 병합
df_10 = pd.merge(df_10, df_10_F, on='회원번호', how='inner')
```

```python
# 주문횟수 이름 변경

df_10.rename(columns = {'주문횟수' : '월간주문횟수'}, inplace = True)
```

```python
# 월간구매금액 기준으로 월간구매금액기준에 HIGH/ LOW만들기
df_10['월간주문횟수기준구분'] = 'LOW'
df_10.loc[df_10['월간주문횟수'] >= 8 , '월간주문횟수기준구분'] = 'HIGH'
df_10
```

```python
df_10['월간주문횟수기준구분'].value_counts()
```

```python
# 두 종목 다 HIGH인 사람들
df_10[(df_10['월간구매금액기준구분'] == 'HIGH') & (df_10['월간주문횟수기준구분'] == 'HIGH')]
# 36858 중 20304명
```

```python
df['주문일시dt'] = pd.to_datetime(df['주문일시'])
df['주문월'] = df['주문일시'].dt.month
```

```python
df[df['주문월'] == 10]['주문일시dt'].value_counts()
```

```python
# 주문일시가 10월인 데이터
df_month_10 = df[df['주문월'] == 10]
df_month_10.info()
```

```python
df_month_10['회원번호'].value_counts()
```

```python
# 10월 회원번호별 주문금액
df1 = df_month_10.groupby('회원번호')['구매금액'].sum().reset_index().sort_values(by = '구매금액' , ascending = False)
```

```python
# 평균 10만원이고 50%가 59156원
```

```python
# 각 월별로 회원번호별 총 구매금액 계산 및 월별 평균 계산
unique_months = df['주문월'].unique()  # 유니크한 월 추출
monthly_order_totals = {}  # 각 월별로 결과를 저장할 딕셔너리

for month in sorted(unique_months):  # 월별로 반복
    # 해당 월의 데이터 필터링
    monthly_data = df[df['주문월'] == month]

    # 회원번호별로 주문금액 합계 계산
    monthly_totals = monthly_data.groupby('회원번호')['구매금액'].sum().reset_index()

    # 월별 평균 계산
    monthly_average = monthly_totals['구매금액'].mean()

    # 결과를 딕셔너리에 저장
    monthly_order_totals[month] = {
        'total_per_member': monthly_totals,
        'average': monthly_average
    }

# 결과 출력
for month, data in monthly_order_totals.items():
    print(f"\n{month}월 회원번호별 총 구매금액:")
    print(data['total_per_member'])
    print(f"{month}월 평균 구매금액: {data['average']:.2f}원")
```

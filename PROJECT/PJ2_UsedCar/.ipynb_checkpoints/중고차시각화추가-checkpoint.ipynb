```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

import seaborn as sns
```

```python
df = pd.read_csv('data_dummies.csv')
df
```

```python
df['Brand_new'] = df['Brand_new'].astype('object')
```

```python
df.info()
```

```python
# 'Brand_new' ì—´ì„ ë¬¸ìžì—´(object) íƒ€ìž…ìœ¼ë¡œ ë³€í™˜
df['Brand_new'] = df['Brand_new'].astype(str)

# 'Brand_new'ì™€ 'Location'ì˜ ë¹ˆë„ ê³„ì‚°
brand_location_counts = df.groupby(['Brand_new', 'Location']).size().reset_index(name='Count')

# Count ê¸°ì¤€ìœ¼ë¡œ ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬
brand_location_counts = brand_location_counts.sort_values(by='Count', ascending=True)

# Barplot ìƒì„±
plt.figure(figsize=(12, 8))
sns.barplot(data=brand_location_counts, x='Location', y='Count', hue='Brand_new', palette='viridis')
plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Distribution of Locations by Brand_new')
plt.xticks(rotation=45)
plt.legend(title='Brand_new')
plt.show()

# => brand 3ê°€ì§€ ë³€ìˆ˜ ìƒì˜ 9ê°œ 3ë“±ê¸‰ 
# (bww) í‰ê·  ê°€ê²© 39749.710225 ì´ìƒ  / 
# ìƒìœ„ 10 ~ 18 /  2ë“±ê¸‰ í‰ê·  ê°€ê²© 39749.710225 ~ í‰ê·  ê°€ê²© (Mahindra) 12453.789140 ê¹Œì§€
# ë‚˜ë¨¸ì§€ 1ë“±ê¸‰ (ford) 12453.789140 ê°€ê²© ì´í•˜ 


# ë¶„í¬ë¥¼ íŒŒì•…í•´ë³´ì•˜ì„ë•Œ ì§€ì—­ë³„ë¡œ 1ë“±ê¸‰ ë¸Œëžœë“œë¥¼ ê°€ì§„ ì°¨ê°€ ê°€ìž¥ ìˆ˜ìš”ê°€ ë†’ë‹¤ => ì¦‰ 1ë“±ê¸‰ ë¸Œëžœë“œë¥¼ ê°€ì§„ ì°¨ë¥¼ íŒë§¤í•˜ëŠ”ê²ƒì´ ìˆ˜ìµì„±ì´ ë†’ë‹¤ 
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

import seaborn as sns
```

```python
df = pd.read_csv('data_dummies.csv')
df
```

```python
df['Brand_new'] = df['Brand_new'].astype('object')
```

```python
df.info()
```

```python
# 'Brand_new' 열을 문자열(object) 타입으로 변환
df['Brand_new'] = df['Brand_new'].astype(str)

# 'Brand_new'와 'Location'의 빈도 계산
brand_location_counts = df.groupby(['Brand_new', 'Location']).size().reset_index(name='Count')

# Count 기준으로 오름차순 정렬
brand_location_counts = brand_location_counts.sort_values(by='Count', ascending=True)

# Barplot 생성
plt.figure(figsize=(12, 8))
sns.barplot(data=brand_location_counts, x='Location', y='Count', hue='Brand_new', palette='viridis')
plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Distribution of Locations by Brand_new')
plt.xticks(rotation=45)
plt.legend(title='Brand_new')
plt.show()

# => brand 3가지 변수 상의 9개 3등급 
# (bww) 평균 가격 39749.710225 이상  / 
# 상위 10 ~ 18 /  2등급 평균 가격 39749.710225 ~ 평균 가격 (Mahindra) 12453.789140 까지
# 나머지 1등급 (ford) 12453.789140 가격 이하 


# 분포를 파악해보았을때 지역별로 1등급 브랜드를 가진 차가 가장 수요가 높다 => 즉 1등급 브랜드를 가진 차를 판매하는것이 수익성이 높다 
```

```python
# 'Fuel_Type'과 'Location'의 빈도 계산
fuel_location_counts = df.groupby(['Fuel_Type', 'Location']).size().reset_index(name='Count')

# Count 기준으로 오름차순 정렬
fuel_location_counts = fuel_location_counts.sort_values('Count')

# Barplot 생성
plt.figure(figsize=(12, 8))
sns.barplot(data=fuel_location_counts, x='Location', y='Count', hue='Fuel_Type', palette='viridis')
plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Distribution of Locations by Fuel Type')
plt.xticks(rotation=45)
plt.legend(title='Fuel Type')
plt.show()


# hyderabad는 다른 원료 보다는 디젤 (경유) 원료 선호하는 경향을 보이고 Mumbai는 디젤 원료 보다는 다른 원료를 선호하는 경향이 보인다

# => 즉 디젤차를 판매하는 경우 다른 지역들보다 hyderabad 지역에 지점을 만들어서 파는것이 좋아보인다 
```

```python
df['Seats'] = df['Seats'].astype(str)

# 'Brand_new'와 'Location'의 빈도 계산
brand_location_counts = df.groupby(['Seats', 'Location']).size().reset_index(name='Count')

# Count 기준으로 오름차순 정렬
brand_location_counts = brand_location_counts.sort_values(by='Count', ascending=True)

# Barplot 생성
plt.figure(figsize=(12, 8))
sns.barplot(data=brand_location_counts, x='Location', y='Count', hue='Seats', palette='viridis')
plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Distribution of Locations by Seats')
plt.xticks(rotation=45)
plt.legend(title='Seats')
plt.show()

# 분포를 살펴봣을때 모든 지역에서 2,4,7인승보다는 other (5인승)이 가장 높게 나타나는 경향을 보인다 
# 따라서 중고차를 판매할때 5인승을 가진 자동차를 판매하는것이 수익성이 높아짐 
```

```python
df['Transmission'] = df['Transmission'].astype(str)

# 'Transmission'과 'Location'의 빈도 계산
fuel_location_counts = df.groupby(['Transmission', 'Location']).size().reset_index(name='Count')

# Count 기준으로 오름차순 정렬
fuel_location_counts = fuel_location_counts.sort_values('Count')

# Barplot 생성
plt.figure(figsize=(12, 8))
sns.barplot(data=fuel_location_counts, x='Location', y='Count', hue='Transmission', palette='viridis')
plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Distribution of Locations by Year')
plt.xticks(rotation=45)
plt.legend(title='Transmission')
plt.show()

# 데이터 조회 결과 모든 지역에서 자동보다는 수동을 선호하는 경향이 보인다  => 즉 수동 기어를 가진 차량 판매하는것이 수익성 향상에 될것으로 보임
```

```python
df['Owner_Type'] = df['Owner_Type'].astype(str)

# 'Transmission'과 'Location'의 빈도 계산
fuel_location_counts = df.groupby(['Owner_Type', 'Location']).size().reset_index(name='Count')

# Count 기준으로 오름차순 정렬
fuel_location_counts = fuel_location_counts.sort_values('Count')

# Barplot 생성
plt.figure(figsize=(12, 8))
sns.barplot(data=fuel_location_counts, x='Location', y='Count', hue='Owner_Type', palette='viridis')
plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Distribution of Locations by Year')
plt.xticks(rotation=45)
plt.legend(title='Owner_Type')
plt.show()

# Owner_Type : 판매차가 몇번째 판매차량인지 확인 
# 데이터 조회 결과 모든 지역에서 첫번째로 판매된 차량을 선호한다  => 여러번 팔린 중고차보다는 한번팔린차량을 선호하는 경향이 많아지기때문에 
# 판매 이력이 한번 밖에 없는 중고차량을 중심적으로 판매하여 수익율을 높인다 
```

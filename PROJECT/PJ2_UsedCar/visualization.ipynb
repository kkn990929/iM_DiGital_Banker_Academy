# 시각화

```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
matplotlib.rc('font', family = 'Malgun Gothic')
matplotlib.rc('axes', unicode_minus=False)
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data_dummies.csv')
```

```python
df.columns
```

```python
sns.displot(df, x = 'Location')
```

```python
sns.displot(df, x = 'Price')
```

```python
sns.displot(df, x = 'Year')
```

```python
sns.displot(df, x = 'Kilometers_Driven')
```

```python
# Location 칼럼의 분포 시각화
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Location', order=df['Location'].value_counts().index)
plt.xticks(rotation=45)
plt.xlabel('Location')
plt.ylabel('Number of Cars')
plt.title('Distribution of Cars by Location')
plt.show()
```

```python
# location_label 칼럼의 분포 시각화
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='location_label')
plt.xlabel('Location Label')
plt.ylabel('Number of Cars')
plt.title('Distribution of Cars by Location Label')
plt.show()
```

```python
# Fuel_Type 칼럼의 분포 시각화
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Fuel_Type', order=df['Fuel_Type'].value_counts().index)
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.title('Distribution of Cars by Fuel Type')
plt.show()
```

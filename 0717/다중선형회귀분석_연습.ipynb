```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns
# 데이터 분할
from sklearn.model_selection import train_test_split
# qq plot , 회귀분석 상수항 추가
from statsmodels.api import qqplot, add_constant
# 선형모델
import statsmodels.formula.api as smf
# 평가함수
from statsmodels.tools.eval_measures import rmse
# 결정계수 평가지표
from sklearn.metrics import r2_score
# 다중공선성 확인
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
matplotlib.rc('font',family = 'Malgun Gothic')
matplotlib.rc('axes', unicode_minus = False)
```

```python
# 데이터 불러오기
df_raw = pd.read_csv('body.csv')
df_raw.head()
```

## 변수간의 경향성 파악 - 그래프 분석

```python
# 산점도 행렬
sns.pairplot(df_raw, y_vars = 'FAT', x_vars=['WEIGHT' , 'NECK', 'CHEST', 'ABDOMEN'] , kind = 'reg');
# 회귀 적합선 추가 kind = 'reg'
```

```python
# 산점도 행렬
sns.pairplot(df_raw);
```

## 변수 간의 경향성 파악 - 상관관계 분석

```python
# 상관관계 분석 
df_raw.corr().round(3)
# FAT - CHEST = 0.703  / FAT - ABDOMEM = 0.813 높은 상관관계
# 이외에도 변수들 간의 다중공성선 확인
```

## 회귀모델 생성 : 설명변수 전체

```python
# 학습용 / 평가용 데이터 분리
df_train, df_test = train_test_split(df_raw, test_size = 0.3, random_state = 4589)
print('학습용 데이터 크기 : {}'.format(df_train.shape))
print('평가용 데이터 크기 : {}'.format(df_test.shape))
# 전체 252개의 데이터를 176/76으로 나눔
```

```python
# 선형 회귀 분석
reg_model = smf.ols(formula ='FAT ~ AGE + WEIGHT + HEIGHT + NECK + CHEST + ABDOMEN + HIP + THIGH + KNEE + ANKLE + BICEPS + FOREARM + WRIST' , data = df_train)
# 적합
reg_result = reg_model.fit()
print(reg_result.summary())

# 모델의 적합성 확인
# R-squared:  0.780
# Adj. R-squared:  0.763
# ANOVA 분석 결과
# Prob (F-statistic): 1.77e-46
# 회귀계수 유의성 검정
# P-value < 0.05 => ABDOMEN , HIP , FOREARM , WRIST
```

```python
# 회귀 모델 생성 : 유의하지 않은 변수 제거 
# 선형 회귀 분석
reg_model = smf.ols(formula ='FAT ~ ABDOMEN + HIP + FOREARM + WRIST' , data = df_train)
# 적합
reg_result = reg_model.fit()
print(reg_result.summary())

# 모델의 적합성 확인 (설명력)
# R-squared: 0.752
# Adj. R-squared: 0.747
# ANOVA 분석 결과
# Prob (F-statistic): 9.86e-51
# 잔차의 정규성 값 검정
# Prob(Omnibus): 0.413
# Durbin-Watson: 1.890
# Jarque-Bera (JB): 1.584
# 정규성은 Prob(Omnibus)나 Durbin_watson 둘 중 하나라도 정규성 기준 통과하면 정규성 검증 통과
```

## 회귀 모델 생성 : 설명변수 간의 다중공선성 확인

```python
# 설명변수 간의 다중공선성 확인
df_raw_x = df_train[['ABDOMEN' , 'HIP' , 'FOREARM' , 'WRIST']]
# statsmodels의 상수항 추가 함수 적용
df_raw_x_const = add_constant(df_raw_x)
# DataFrame을 저장
df_vif = pd.DataFrame()
df_vif['Variable'] = df_raw_x_const.columns
df_vif['VIF'] = [variance_inflation_factor(df_raw_x_const.values, i) for i in range(df_raw_x_const.shape[1])]
# df_raw_x_const의 인덱스 값 1(5)만 큼 반복해서 각 values별로 VIF값 구하기
# VIF 정렬
df_vif.sort_values('VIF', inplace = True)
df_vif.round(3)

# VIF 값이 1~10미만이면 다중공선성이 문제가 되지 않음
```

## 오차의 기본가정 검토

### 등분산성 검정

```python
# 잔차의 등분산성 : X-예측값, Y-잔차
plt.scatter(reg_result.fittedvalues, reg_result.resid);
# fittedvalues는 모델이 예측한 종속 변수의 값
# resid는 잔차(residuals)를 의미
plt.xlabel('fitted value')
plt.ylabel('residual')
plt.axhline(y = 0, linestyle = '--');
# print(reg_result.fittedvalues)
# 잔차가 y = 0를 기준으로 Random하게 산포되어 있으므로, 등분산성 만족
```

### 독립성 검정

```python
# 잔차의 독립성 검정 X - 데이터 순서 / y - 잔차
plt.scatter(df_train.index, reg_result.resid, marker = 'X')
plt.xlabel('DATA')
plt.ylabel('RESIDUAL')
plt.axhline(y = 0, linestyle = '-') # CL
plt.axhline(y = 3*np.std(reg_result.resid), linestyle = '--') # 3IQR (UCL)
plt.axhline(y = -3*np.std(reg_result.resid), linestyle = '--'); # -3IQR (LCL)
# print(reg_result.resid)
print(df_train.index)
# 경향성 없고, 관리기준 초과 데이터 없으니 독립성 만족
```

### 정규성검정

```python
# 잔차 정규성 검정 Q-Q plot
qqplot(reg_result.resid, line = 'q');
```

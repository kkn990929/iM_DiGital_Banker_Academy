## 1. 5의 배수로 구성된 9개의 값을 생성한 후, Numpy Matrix 3x3 행렬로 만든 후, 첫번째 행의 합과 평균을 구하는 함수를 생성

|  |  |  |
|--|--|--|
|5 |10|15|
|20|25|30|
|35|40|45|

```python
import numpy as np
def matrix_cal():
    #5의 배수로 구성된 9개 값
    values = np.arange(5, 46, 5)
    #3*3 행렬
    matrix = values.reshape(3,3)
    # matrix
    #첫 행 추출
    row1 = matrix[0]
    #첫 행의 합
    row1_sum = row1.sum()
    #첫 행의 평균
    row1_mean = row1.mean()
    return matrix, row1_sum, row1_mean

matrix_cal()
```

```python
def func1():
    list1 = [x for x in range(1, 50) if x % 5 == 0]
    data1 = np.array(list1)
    data2 = data1.reshape(3,3)
    return data2[0].sum() ,data2[0].mean()
     

func1()
```

## 8. 본사에서 매출관리를 위해 연체가 높은 고객들을 확인하여 관리하고자 한다. 연체건수가 높은 순서대로 정렬하고, 상위 100명의 고객의 연령(Age)의 최대값과 최소값을 확인, 신용등급(Credit_Rank)의 평균을 계산 

```python
#연체 건수가 높은 순서로 정렬
df33 = df1.sort_values( by = 'Overdue_count', ascending = False).head(100)
#최대값
max_Age = df33['Age'].max()
#최소값
min_Age = df33['Age'].min()
#신용등급의 평균
mean_Credit_Rank=df33['Credit_Rank'].mean()

print(max_Age,min_Age,mean_Credit_Rank)
```

## 9. 매출관리 부서 A씨는 매주 해당 데이터를 받아 계약금액이 높은 1000개의 계약명단의 평균값을 계산하고 있다. 이 업무를 자동화 하기 위해, 월랜탈비용이 높은 1000갸의 데이터를 추출하여, 해당 데이터의 월랜탈비용의 평균을 계산하는 함수를 만드시오. 

```python
#Amount_Month가 높은 1000개의 명단 추출 and 평균 계산함수
def cal_AmountMonth():
    #계약금액이 높은 1000개의 데이터 추출
    df10 = df1.sort_values( by = ['Amount_Month'] , ascending = False).head(1000)
    #1000개의 데이터의 평균
    Amount_month_mean = df10['Amount_Month'].mean()

    return Amount_month_mean

cal_AmountMonth()

```

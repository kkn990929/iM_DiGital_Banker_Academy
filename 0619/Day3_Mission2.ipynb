## 2. 사업자와 일반 고객의 사용한 적립금액의 합과 평균을 계산한 후 각각의 차를 계산하시오

```python
#주문 시 회원등급
# 사용한 적립금액
data1 = df9[df9['주문 시 회원등급'] == '사업자']
data2 = df9[df9['주문 시 회원등급'] == '일반']

sum_BM = data1['사용한 적립금액'].sum()
sum_R = data2['사용한 적립금액'].sum()

mean_BM = data1['사용한 적립금액'].mean()
mean_R = data2['사용한 적립금액'].mean()

#각각의 합의 차
diff_sum = sum_BM - sum_R
diff_mean = mean_BM - mean_R


print(diff_sum, diff_mean)
```

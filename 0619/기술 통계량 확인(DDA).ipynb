# 기술 통게량 확인 (DDA)
- 데이터의 통계적 특성을 확인
## 연속형
- 대표값 ( 평균/ 중앙값) : 해당 데이터를 대표하는 값
- 산포 (분산 / 표준편차) : 대표값에 대해 데이터가 얼마나 정확한지 나타냄. 데이터의 신뢰도 확인
- 분포의 모양 : 분포의 모양에 따라 가설검정 기법/ 회귀분석 방법이 달라짐. 가설 검정을 통해 정규성 확인

```python
import pandas as pd
df9 = pd.read_csv('09_Data.csv', encoding = 'cp949')
# .cp949 / euc-kr / utf-16 / utf-8 / ascii
df9
```

## 1. Unnamed:0와 카드사 컬럼을 제외하여 df10으로 선언한 후, df10의 컬럼명을 확인하시오

```python
df10 = df9.drop(columns = ['Unnamed: 0', '카드사'])
#새로운 변수 선언 or inplace 사용
df10
```

## 2. df10의 구조와 타입을 확인하시오

```python
df10.info()
```

## 3. df10의 결측치를 확인한 후, 결측치가 있는 행들을 제거하여 df10_clean으로 선언하고, df10_clean의 결측치 유무를 확인하시오.

```python
df10_clean = df10.dropna()
df10_clean.isnull().sum()
```

## 5. df10_clean의 주문 시 회원등급의 항목과 빈도수를 확인하시오

```python

print(df10_clean['주문 시 회원등급'].unique(),df10_clean['주문 시 회원등급'].value_counts())
```

## 6. df10_clean의 중 '주문시 회원 등급' 별 주문한 수량 (수량)의 합과 평균을 확인하시오

```python
p1 = df10_clean.pivot_table(index = '주문 시 회원등급' , values = '수량' , aggfunc = ['sum','mean'])
p1.columns = p1.columns.droplevel(1)
p1.columns
```

# 날짜 데이터
- 컴퓨터가 숫자로 된 날짜를 인식할 수 있게끔 함

```python
import pandas as pd
df1 = pd.read_csv('01_Data.csv')
```

```python
df1['Datetime_dt'] = pd.to_datetime(df1['Datetime'])
#Datetime_dt라는 칼럼을 추가하고 내용은 to_datetime으로 추가
df1.head()
```

```python
df1[['Datetime' , 'Datetime_dt']]
```

```python
df1['Year'] = df1['Datetime_dt'].dt.year
df1['Month'] = df1['Datetime_dt'].dt.month
df1['Day'] = df1['Datetime_dt'].dt.day

df1['Day_of_week'] = df1['Datetime_dt'].dt.day_name() #요일 추출
df1['week'] = df1['Datetime_dt'].dt.isocalendar().week #주차 추출

df1.head()
```

# 파생변수 생성

```python
df1.head()
```

```python
# 월 렌탈 비용과 계약 기간을 곱하여 총 계약 금액을 구하자
df1['Amount_Month'] * df1['Term']
```

```python
df1['Total_Amount'] = df1['Amount_Month'] * df1['Term'] #total amount 라는 칼럼을 만듬
df1.head()
```

```python
#연령대
cond1 = (df1['Age'] >= 70)
cond2 = (df1['Age'] >= 60)
cond3 = (df1['Age'] >= 50)
cond4 = (df1['Age'] >= 40)
cond5 = (df1['Age'] < 40)

#조건을 넣어서 만들기 .loc[조건, 새로생성할 칼럼] = '부여할 값'
df1.loc[cond5, '연령대'] = '40대 미만'
df1.loc[cond4, '연령대'] = '40대 '
df1.loc[cond3, '연령대'] = '50대 '
df1.loc[cond2, '연령대'] = '60대 '
df1.loc[cond1, '연령대'] = '70대 이상'

df1[['Age','연령대']]
```

```python
df1['연령대'].value_counts()
```

```python
# 노가다 그만

def agefunc(row):
    if row >= 70:
        return '70대 이상'
    elif row >= 60:
        return '60대'
    elif row >= 50:
        return '50대'
    elif row >= 40:
        return '40대'
    else:
        return '40대 미만'


df1['Age'].apply(agefunc)
        
```

```python
df1['연령대2'] = df1['Age'].apply(agefunc)
df1.head()
```

```python
# 더 쉽게!

def agefunc2(i):
    return str(i)[0] + '0대'

df1['연령대3'] = df1['Age'].apply(agefunc2)
df1['연령대3'].value_counts()
```

```python
def func3(x):
    if x >= 100:
        return '100대 이상'
    else:
        return str(x)[0] + '0대'

df1['연령대4'] = df1['Age'].apply(func3)
df1['연령대4'].value_counts()
```

# 병합 merge
- key column : 병합시 기준이 되는 칼럼

```python
data1 = { '이름' : ['신짱구', '신짱아' , '신형만'] , '성별' : ['남','여','남'] , '나이' : [5, 3, 30]}

f1=pd.DataFrame(data1)

data2 = { '이름' : ['신짱구' , '신짱아' , '철수' , '유리'] , '부서' : ['A', 'B', 'C','D'] , '직급' : ['사원', '사원', '주임', '본부장']}

f2=pd.DataFrame(data2)
```

```python
pd.merge(f1, f2, on='이름' , how = 'inner') #innerjoin 키 col 기준 양 쪽 교집합
```

```python
f1
```

```python
f2
```

```python
pd.merge(f1, f2, on='이름' , how = 'outer') #outerjoin 키 컬럼 기준 양 쪽 합집합 + 결측값 많아짐
```

```python
#left / right join
pd.merge(f1, f2, on='이름' , how = 'left')
# left에 존재하는 이름 값을 기준으로 right를 붙임
```

```python
pd.merge(f1, f2, on='이름' , how = 'right')
#right에 존재하는 이름 값 기준으로 left를 붙임
```

# 파생변수
- quantile()

```python
df1.head()
```

```python
df1['총구매금액'].quantile(0.85)
```

```python
df1.describe()
```

```python
dc = df1['총구매금액'].describe()
dc
```

```python
def rankfunc(row):
    if row >= dc['75%']:
        return 'A'
    elif row >= dc['50%']:
        return 'B'
    elif row >= dc['25%']:
        return 'C'
    else:
        return 'D'

df1['등급'] = df1['총구매금액'].apply(rankfunc)
df1[['총구매금액','등급']]
    
```

```python
df1['등급'].value_counts()
```

```python
# 총ㄱ 구매금액이 85% 이상이면 'viP'
# 그렇지 않은 경우 일바
# 
qun85 = df1['총구매금액'].quantile(0.85)
def totalfunc(row):
    if row >= qun85:
        return 'VIP'
    else:
        return '일반'

df1['등급2'] = df1['총구매금액'].apply(totalfunc)

df1['등급2'].value_counts()

print(df1['등급2'].value_counts())
print(qun85)
```

# 파생변수 생성
- 기존의 col을 사용하여 새로운 col생성

```python
# 총 계약 금액
# 월 렌탈 금액 * 계약 기간
df1['Total']  =df1['Amount_Month'] * df1['Term']
df1[['Amount_Month', 'Term', 'Total']]
```

```python
# 연령대
cond1 = (df1['Age'] >= 60)

df1.loc[cond1 , '연령대']  = '60대 이상'
# df.loc[조건 , 생성할 col] = 부여할 값
df1.loc[~cond1, '연령대'] = '60대 이하'

df1.head()
```

```python
# 함수 만들어서 파생변수 만들기
def func1(row):
    if pd.isnull(row):
        return '미확인'
    elif row >= 100:
        return '100세 이상'
    else:
        return str(row)[0] + '0대'

df1['연령대2'] = df1['Age'].apply(func1)
df1['연령대2'].value_counts()
```

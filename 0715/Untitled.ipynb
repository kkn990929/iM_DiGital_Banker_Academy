```python
import pandas as pd
```

```python
left = pd.DataFrame({'key1' : ['K0', 'K1' , 'K2', 'K3'] , 'key2' : ['K0', 'K1', 'K0', 'K1'], 'A':['A0', 'A1', 'A2', 'A3'] , 'B' :['B0', 'B1','B2','B3']})
left
```

```python
right = pd.DataFrame({'key1' : ['K0', 'K1' , 'K2', 'K3'] , 'key2' : ['K0', 'K0', 'K0', 'K0'], 'C':['C0', 'C1', 'C2', 'C3'] , 'D' :['D0', 'D1','D2','D3']})
right
```

```python
result = pd.merge(left, right, on=['key1','key2'])
result
```

```python
result1 = pd.merge(left, right, on=['key1', 'key2'], how = 'left', indicator= True)
result1
```

```python
df = pd.DataFrame({'foo': ['one','one','one','two','two','two'], 'bar' : ['A','B','C','A','B','C'], 'barz' : [1,2,3,4,5,6], 'zoo': ['x','y','z','q','w','t']})
df
```

```python
df.pivot(index= 'foo', columns = 'bar', values = 'barz')
```

```python
df = pd.DataFrame({'factor1': ['A','A','A','A','A','B','B','B','B'], 'factor2' : ['a','b','c','a','c','a','b','c','b',], 
              'value' :[11,12,13,11,12,13,11,11,12]})
df
```

```python
pd.crosstab(df.factor1, df.factor2)
```

```python
pd.crosstab(df.factor1,df.factor2, margins =True, normalize = True)
# margins는 총 합계량, normalize는 빈도산출
```

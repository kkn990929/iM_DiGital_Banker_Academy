```python
import pandas as pd
import numpy as np
```

```python
pop_seoul = pd.read_csv('서울인구통계.txt', sep='\t', thousands = ',' , skiprows = 2, encoding = "utf-8" ) 
pop_seoul.head()
```

```python
pop1 = pop_seoul[['자치구' , '계', '계.1', '계.2','65세이상고령자']]
pop_seoul = pop1.drop([0])
pop_seoul.head()
```

```python
pop_seoul.rename(columns = {pop_seoul.columns[0]:'구별',pop_seoul.columns[1]:'인구수' , pop_seoul.columns[2]:'한국인' , pop_seoul.columns[3]:'외국인', pop_seoul.columns[4]:'고령자'},inplace = True)
pop_seoul.head()
```

```python
cctv_seoul = pd.read_excel('서울시CCTV.xlsx', skiprows = 2, names = ['기관명', '소계', '2011년 이전' , '2012년' , '2013년' , '2014년','2015년', '2016년', '2017년'  , '2018년' , '2019년', '2020년' , '2021년'])
cctv_seoul.head()
```

```python
cctv_seoul.rename(columns = {cctv_seoul.columns[0] : '구별'},inplace = True)
cctv_seoul.rename(columns = {cctv_seoul.columns[1] : 'CCTV수'},inplace = True)
cctv_seoul.head()
```

```python
cctv_seoul.sort_values(by = 'CCTV수',ascending = True).head(10)
```

```python
cctv_seoul.sort_values(by = 'CCTV수',ascending = False).head(10)
```

```python
cctv_seoul.isnull().sum()
```

```python
cctv_seoul['최근증가율'] = (cctv_seoul['2021년'] + cctv_seoul['2020년'] + cctv_seoul['2019년']) /  (cctv_seoul['CCTV수'] - (cctv_seoul['2021년'] + cctv_seoul['2020년'] + cctv_seoul['2019년'])) * 100
```

```python
cctv_seoul.head()
```

```python
cctv_seoul.sort_values(by = '최근증가율' , ascending = False).head(5)
```

```python
pop_seoul['외국인 비율'] = pop_seoul['외국인']/pop_seoul['인구수']*100
pop_seoul['고령자 비율'] = pop_seoul['고령자']/pop_seoul['인구수']*100
pop_seoul.head()
```

```python
# 서울시 인구 데이터 확인
pop_seoul.sort_values(by = '외국인 비율' , ascending = False).head()
```

```python
pop_seoul.sort_values(by = '고령자 비율' , ascending = False).head()
```

```python
cctv_seoul['구별'] = cctv_seoul['구별'].str.replace(' ','')
```

```python
result = pd.merge(cctv_seoul,pop_seoul, on = '구별')
result.head()
```

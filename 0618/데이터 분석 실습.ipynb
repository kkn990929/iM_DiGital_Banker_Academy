# 데이터 구조, 타입 확인
- index, column, value 확인
- index = 데이터 갯수 확인
      - 데이터 갯수에 따라 분석 방법 달라짐
- column = 데이터의 열 / 데이터의 항목
      - 각 항목별 데이터 타입 확인
      - 연속형 int/float
      - 범주형 obj
      - 순서형 날짜/시간데이터(datetime)
- value : 값 확인
      - 데이터의 형식 확인
      - 자료 설명서/DB설계서/변수정의서 확인
      - 결측값(수집 중 오류나 누락으로 빈 값(none/NaN/Null/Na)) 확인

# 데이터 정렬
- .sort_values(by ~)
- (, ascending = False)

```python
#default는 오름차순 
df1.sort_values(by = 'Amount_Month')
```

```python
#내림차순 
# ascending = False
df1.sort_values(by = 'Amount_Month',ascending = False)
```

```python
# 여러항목 기준 정렬
# Amount_Month와 overdue_count 기준 
# 월 구매금액이 높으면서 연체횟수가 낮은 고객 찾기
df2 = df1.sort_values( by = ['Amount_Month','Overdue_count'], ascending = [False, True])
```

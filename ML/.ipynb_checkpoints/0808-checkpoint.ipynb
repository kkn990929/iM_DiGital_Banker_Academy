# 주성분분석
- 차원축소
- 원본데이터의 분산 = 공분산 행렬

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris_df = pd.DataFrame(data = iris.data, columns = iris.feature_names)
iris_df['target'] = iris.target
markers = ['*','s','o']

for i, marker in enumerate(markers):
    x_val = iris_df[iris_df['target'] == i]['sepal length (cm)']
    y_val = iris_df[iris_df['target'] == i]['sepal width (cm)']
    plt.scatter(x_val, y_val, marker = marker, label = iris.target_names[i])

plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
iris_df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(iris_df)

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(data_scaled)
df_pca = pca.transform(data_scaled)

df_pca = pd.DataFrame(data = df_pca)
df_pca['target'] = iris.target

markers = ['*','s','o']

for i, marker in enumerate(markers):
    x_val = df_pca[df_pca['target'] == i][0]
    y_val = df_pca[df_pca['target'] == i][1]
    plt.scatter(x_val, y_val, marker = marker, label = iris.target_names[i])

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()

print(pca.explained_variance_ratio)
    
```

# 예측 검증 확인

### MAE (Mean Absolute Error)
- 실제 값과 측정(예측) 값과의 차이
- 모든 절대 오차의 평균
- 예측값과 실제값의 거리

```python
df= pd.read_csv("C:/Users/campus4D037/iM DiGital Bank/ML/datas/Prediction_Sample.csv", encoding = 'cp949')
df = df.dropna()
x = df[['X1','X2']]
y = df['Y1']
x_train, x_test, y_train, y_test = sms.train_test_split(x,y,train_size = 0.8, test_size = 0.2)
model = slm.LinearRegression()
model.fit(x_train,y_train)

predict_x = model.predict(x)

import sklearn.metrics as slt

MEA_score = slt.mean_absolute_error(predict_x,y)

print('MEA : ', MEA_score)

# 평균절대오차거리 15
```

### RMSE
- 예측 모델에서 예측한 값과 실제 값 사이의 평균 차이를 측정
- 예측 모델이 목표 값(정확도)을 얼마나 잘 예측할 수 있는지 추정
- 실제값과 분산과 예측값의 분산을 상대적으로 구한것 -> R^
 

```python
MEA_score = slt.mean_absolute_error(predict_x,y)
MSE_score = slt.mean_squared_error(predict_x,y)

RMSE_score = MSE_score ** 0.5

print('MAE : ', MEA_score, 'MSE : ',MSE_score, 'RMSE : ' , RMSE_score)
      
```

### R-squared
- 실제값과 분산과 예측값의 분산을 상대적으로 구한것 -> R^


```python
MEA_score = slt.mean_absolute_error(predict_x,y)
MSE_score = slt.mean_squared_error(predict_x,y)

RMSE_score = MSE_score ** 0.5
print('------------평균기반 검증-------------')
print('MAE : ', MEA_score, 'MSE : ',MSE_score, 'RMSE : ' , RMSE_score)

print('-------------상대적 분산 기반 검증------------')

R2_score = slt.r2_score(predict_x,y)

print('R2_score :' ,R2_score)
```

# SVM(Support Vector Machchine)


```python
import pandas as pd
import sklearn.model_selection as sms
import sklearn.svm as skv
df = pd.read_csv("C:/Users/campus4D037/iM DiGital Bank/ML/datas/SVM_Sample.csv")
df=df.dropna()
```

```python
x = df[['X1','X2']]
y = df['Y1']

x_train , x_test, y_train, y_test = sms.train_test_split(x,y,train_size = 0.8, test_size = 0.2)
```

```python
SVM_model = skv.SVC(kernel = 'linear', C=1)
SVM_model.fit(x_train, y_train)
Train_score = SVM_model.score(x_train,y_train)
Test_score = SVM_model.score(x_test,y_test)

print('SVM-train score : ', Train_score ,'SVM-TEST score :' , Test_score) 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

# 2차원 데이터 생성 (make_blobs)
X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=6)

# SVM 모델 생성 및 훈련
model = SVC(kernel='linear', C=1)
model.fit(X, y)


def plot_svm_decision_boundary(X, y, model):
    plt.figure(figsize=(10, 6))
    
    # 전체 공간을 평가하기 위한 그리드 생성
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))
    
    # 모델의 예측을 통해 결정 경계 그리기
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # 결정 경계 시각화
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=50, edgecolor='k', marker='o')
    
    # 서포트 벡터 시각화
    plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
                s=100, facecolors='none', edgecolors='k', marker='o', label='Support Vectors')
    
    plt.title('SVM Decision Boundary and Support Vectors')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.show()

# 시각화 호출
plot_svm_decision_boundary(X, y, model)
```

```python
# 3차원 군집화
k_means_model = skc.KMeans(n_clusters=4, random_state=42)
k_means_model.fit(df_kmeans)
Re = k_means_model.predict(df_kmeans)

# 실루엣 점수 계산
s_score = skm.silhouette_score(df_kmeans, Re)
print("Silhouette Score:", s_score)

# 3D 시각화를 위한 데이터 준비
x = df_kmeans['Kilometers_Driven']
y = df_kmeans['Price']
z = df_kmeans['Seats']

# 3D 플롯 생성
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c=Re, cmap='viridis', s=50)

# 색상바 추가
cbar = plt.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label('Cluster Label')

# 축 레이블 설정
ax.set_xlabel('Kilometers Driven')
ax.set_ylabel('Price')
ax.set_zlabel('Seats')

# 플롯 제목 설정
ax.set_title('3D K-Means Clustering')

# 플롯 표시
plt.show()
```

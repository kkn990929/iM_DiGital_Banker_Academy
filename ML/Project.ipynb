```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

# '지역이 A'인 데이터 필터링
df_A = df[df['위치'] == 'A']

# 꺾은선 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(df_A['일자'], df_A['현재발전출력'], marker='o', linestyle='-', color='b')
plt.title('현재발전출력 over Time for 위치 A')
plt.xlabel('일자')
plt.ylabel('현재발전출력')
plt.grid(True)
plt.show()
```

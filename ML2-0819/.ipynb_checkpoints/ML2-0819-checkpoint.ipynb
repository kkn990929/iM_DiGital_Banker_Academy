## 인공신경망

```python
# 성적과 공부시간에 대한 데이터
x1 = [2,4,6,8] # 공부시간
x2 = [1,2,1,3] # 문제집 푼 횟수
y =  [20,55,45,80] # 성적
```

```python
# 컴퓨터한테 자료를 줄때는 리스트가 아닌 행렬로 만들어야
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array(x1)
x2= np.array(x2)
y = np.array(y)
```

```python
# 오차값이 최대일때의(axx +b 곡선의 a,b값이 0일때) a값에 대해 편미분(학습)을 진행하여 a값을 축소시키고 a 갑이 최종적으로 0일 때 -> 학습완료
```

```python
# 실습
```

```python
a1 = 0
a2 = 0
b = 0

lr = 0.001

for i in range(3000):
    y_hat = a1 * x1 + a2 * x2 + b # 예측선 
    error = y_hat - y # 오차구하기
    a1_diff = sum(2 * error * x1) # a1에 대한 편미분값
    a2_diff = sum(2 * error * x2) # A2에 대한 편미분값 
    b_diff = sum(2 * error * 1) # b에 대한 미분값
    a1 = a1 - a1_diff * lr
    a2 = a2 - a2_diff * lr
    b = b - b_diff * lr
    
    # print(f"{i+1}회 학습/ a1 기울기 : {a1_diff : .2f}, b 기울기 :{b_diff:.2f}")
    print(f'{i+1}회 학습 / a1 기울기 : {a1_diff:.2f}, a2 기울기 : {a2_diff:.2f}, b 기울기 : {b_diff:.2f}')
```

```python

plt.scatter(x1,y)
plt.plot(x1,a1 * x1 + a2 * x2 + b, 'red')
plt.grid()
plt.show()
```

```python
a1 * 3 + a2 * 2 +b # 예측값
```

```python
# 사인함수와 가장 오차가 작은 직선 찾기
```

```python
x = np.arange(-3,3,0.01)
y = np.sin(x)

a = 0
b = 0
c = 0
```

```python
plt.plot(x,y)
plt.plot(x,a * x + b, 'red')
plt.grid()
plt.show()
```

```python
# 사인함수와 가장 오차가 작은 직선 찾기
lr = 0.000001

for i in range(5000):
    y_hat = a * x + b # 예측선 
    
    error = y_hat - y # 오차구하기
    
    a_diff = sum(2 * error * x) # a1에 대한 편미분값 
    b_diff = sum(2 * error * 1) # b에 대한 미분값
    # c_diff = sum(
    a = a - a1_diff * lr 
    b = b - b_diff * lr
    
    # print(f"{i+1}회 학습/ a 기울기 : {a1_diff : .2f}, b 기울기 :{b_diff:.2f}")
    print(f'{i+1}회 학습 / a 기울기 : {a_diff:.2f}, b 기울기 : {b_diff:.2f}')
```

```python
plt.scatter(x,y)
plt.plot(x,a * x + b, 'red')
plt.grid()
plt.show()
```

```python


a = 0
b = 0
c = 0
lr = 0.00005

for i in range(5000):
    y_hat = a * x ** 2 + b * x + c
    error = y_hat - y
    
    a_diff = sum(2 * error * x ** 2)
    b_diff = sum(2 * error * x)
    c_diff = sum(2 * error * 1)
    
    a -= a_diff * lr
    b -= b_diff * lr
    c -= c_diff * lr
    
    print(f"{a_diff:.2f}, {b_diff:.2f}, {c_diff:.2f}")
```

```python
plt.scatter(x,y)
plt.plot(x,a * x ** 2 + b * x + c, 'red') # 이차함수로 입력했으나 컴퓨터가 스스로 2차 보다는 일차가 낫다고 판단후 이차항 계수 제거
plt.grid()
plt.show()
```

```python

# 3차 함수로도 그리기
a = 0
b = 0
c = 0
d = 0
lr = 0.000005

for i in range(5000):
    y_hat = a * x ** 3 + b * x ** 2  + c * x + d
    error = y_hat - y
    
    a_diff = sum(2 * error * x ** 3)
    b_diff = sum(2 * error * x ** 2)
    c_diff = sum(2 * error * x)
    d_diff = sum(2 * error * 1)
    a -= a_diff * lr
    b -= b_diff * lr
    c -= c_diff * lr
    d -= d_diff * lr
    print(f"{a_diff:.2f}, {b_diff:.2f}, {c_diff:.2f}")
```

```python
plt.scatter(x,y)
plt.plot(x,a * x ** 3 + b * x ** 2  + c * x + d, 'red') # 이차함수로 입력했으나 컴퓨터가 스스로 2차 보다는 일차가 낫다고 판단후 이차항 계수 제거
plt.grid()
plt.show()
```

## K-NN 지도 학습

```python
# length, weight
x = fish[['length','weight']]
y = fish['target']

# 행렬 처리
x = np.array(fish[['length','weight']])
y = np.array(fish['target'])

# K-NN 호출
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier() # 모델생성 

model.fit(x,y) # 학습


```

```python
# 길이가 40 무게가 600이면?

a = [40,600]

model.predict([a])
```

```python
# 훈련데이터 , 시험데이터 생성
# 훈련데이터, 시험데이터 내의 비율도 맞춰줘야
# 각 세트내의 생선의 비율도 같아야함
from sklearn.model_selection import train_test_split

train_x , test_x , train_y, test_y =train_test_split(x,y,test_size = 0.2)
```

```python
plt.scatter(train_x[:,0] , train_x[:,1])
plt.scatter(test_x[:,0] , test_x[:,1])
plt.legend('train','test')
plt.show()
```

```python
model = KNeighborsClassifier() # 모델생성 

model.fit(train_x,train_y)
```

```python
model.predict(test_x)# 예측값
```

```python
test_y # 실제값 
```

```python
model.score(test_x,test_y) # 점수
```

```python
a = [25,150]

model.predict([a])

# 틀림
```

```python
dis , idx = model.kneighbors([a])
idx

# a(25,150)근처에 점들 추출
```

```python
plt.scatter(train_x[:,0] , train_x[:,1])
plt.scatter(25,150,marker = 'x')
plt.scatter(train_x[idx,0] , train_x[idx,1],marker = 'o')
plt.show()

# ML이 보기에는 빙어가 더 가까움
```

```python
# 정규분포 만들기 (데이터 표준화)
mean = train_x.mean(axis = 0) # 길이, 무게 평균 
std = train_x.std(axis = 0) # 길이, 무게 표준편차
```

```python
# 길이 - 길이평균 / std
# 무게 - 무게평균 / std
train_scaled = (train_x - mean) / std

test_scaled = (test_x - mean) / std

# 훈련데이터의 평균으로 빼야함 test는 모르는값이라고 가정하기 떼문에 평균 구할 수 없음
```

```python
plt.scatter(train_scaled[:,0] , train_scaled[:,1])
# 데이터 왜곡 없음, 근데 x축 y축 범위만 바뀜
```

```python
model = KNeighborsClassifier() # 모델생성 

model.fit(train_scaled,train_y) # 학습
model.score(test_scaled,test_y) # 점수

# 잘 학습된 모델
```

```python
a = ([25,150] - mean) / std

model.predict([a])

# 예측 성공 
```

```python
# 검증
model.kneighbors([a])
```

```python
dis , idx = model.kneighbors([a])
idx

# a(25,150)근처에 점들 추출
```

```python
plt.scatter(train_scaled[:,0] , train_scaled[:,1])
plt.scatter(a[0],a[1],marker = '*')
plt.scatter(train_scaled[idx,0] , train_scaled[idx,1],marker = 'D')
plt.show()
```

```python
# 정규분포 만들기 (데이터 표준화)
mean = train_x.mean(axis = 0) # 길이, 무게 평균 
std = train_x.std(axis = 0) # 길이, 무게 표준편차


# 길이 - 길이평균 / std
# 무게 - 무게평균 / std
train_scaled = (train_x - mean) / std

test_scaled = (test_x - mean) / std

# 훈련데이터의 평균으로 빼야함 test는 모르는값이라고 가정하기 떼문에 평균 구할 수 없음
```

```python
model = KNeighborsClassifier()
model.fit(train_scaled, train_y)


# 최적의 이웃 찾기
box = []
for i in range(1,31):
    model.n_neighbors = i
    score = model.score(test_scaled, test_y)
    box.append(score)
```

```python
plt.plot(range(1,31),box)
plt.show()


model.n_neighbors = 1

# 반복해서 이웃을 찾을떄마다 점수가 떨어짐
```

```python
model.score(test_scaled,test_y)
```

### 비정형 데이터 -> 딥러닝
### 정형데이터
- 데이터가 균형하고 많다 -> RF ( 다수결 학습 = 많을수록 유리)
- 데이터가 불균형하고 많다 -> XGboost (오차가 많을 수록 잘 학습)
- 데이터가 불균형한데 적다 -> SVM
- 대부분의 정형데이터는 트리기반 모델이 잘 나옴(결정트리, RF, Adaboost, XGboost)
    1. 쉽다 -> 데이터표준화 X -> 연산 X
    2. 강력하다 -> 성능 검증 ok except 결정트리
    3. 데이터 설명력 있음

## 결정트리

```python
iris
```

```python
iris = sns.load_dataset('iris')

x = iris.iloc[:,:-1].to_numpy()
y = iris['species'].to_numpy()

train_x, test_x, train_y, test_y = train_test_split(x, y) # 75 : 25
```

```python
# 결정트리 모델
# 트리모델은 표준화 XXXX
from sklearn.tree import DecisionTreeClassifier, plot_tree

model = DecisionTreeClassifier(max_depth = 4) # max_depth로 모델의 학습 조정
model.fit(train_x, train_y)
model.score(test_x, test_y)
```

```python
plt.figure(figsize = (10, 10))
plot_tree(model, filled = True, feature_names=['sl', 'sw', 'pl', 'pw'])
plt.show()
```

- 훈련 >>> 시험 => 과대적합
- 훈련 > 시험 => 잘 학습
- 훈련 < 시험 => 과소적합
- 둘 다 작게 => 과소적합

```python
# 모델의 과대적합확인
print(model.score(train_x, train_y)) # 훈련점수
print(model.score(test_x, test_y)) # 시험점수
```
